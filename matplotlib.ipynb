# Bar Plot

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cols = ['date', 'close', 'high', 'low', 'open']
df = pd.read_csv("AAPL.csv", index_col="date", usecols=cols, parse_dates=["date"])
df
```

```python
years = df.resample("a").mean().round(2)
years.index = years.index.year
years
```

```python
years.plot(y="high", kind="bar", xlabel="Year")
```

```python
years.plot(y="high", kind="barh", xlabel="Year")
```

```python
years.plot(y=["high", "low"], kind="bar", color=["green", "red"], xlabel="Year")
```

# Pie Charts

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("TopRichestInWorld.csv")
df
```

```python
industries = df.Industry.value_counts()
industries
```

```python
plt.pie(x=industries, labels=industries.index)
plt.show()
```

```python
labels = industries.index[:10]
values = industries.values[:10]
plt.pie(x=values, labels=labels, autopct="%.2f%%")
plt.show()
```

```python
plt.pie(x=values, labels=labels, colors = ["darkblue", "skyblue"], autopct="%.0f")
plt.show()
```

# Styles

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Pokemon.csv", index_col="#")
df
```

```python
poke_groups = df.groupby(by="Type 1")
max_attack = poke_groups["Attack"].max()
```

```python
max_attack.sort_values().plot(kind="barh", color=("skyblue", "crimson"))
```

```python
plt.style.available
```

```python
plt.style.use("grayscale")
max_attack.plot(kind="barh")
```

```python
plt.style.use("dark_background")
max_attack.plot(kind="barh")
```

```python
for style in plt.style.available:
    plt.style.use(style)
    print(style)
    max_attack.plot(kind="barh", figsize=(4,3), fontsize=6)
    plt.show()
```

# Color by Value

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("gap.csv", usecols=["Country", "City", "AQI Value", "AQI Category"])
df
```

```python
df.sort_values(by="AQI Value", inplace=True)
df
```

```python
countries = df.groupby("Country")
```

```python
average = countries.mean(numeric_only=True).round(2)
average
```

```python
average.sort_values(by="AQI Value", inplace=True)
average
```

```python
def value2color(value):
    if value<50:
        return "forestgreen"
    elif value<100:
        return "orange"
    else:
        return "crimson"
```

```python
colors = average["AQI Value"].apply(value2color)
colors
```

```python
average.plot(y="AQI Value", color=colors, kind="barh", figsize=(10,40))
```

```python
turkey = countries.get_group("Turkey")
turkey
```

```python
colors = turkey["AQI Value"].apply(value2color)
colors
```

```python
turkey.plot(y="AQI Value", x="City", kind="barh", figsize=(10,40), color=colors)
plt.show()
```

```python
from matplotlib.patches import Patch
```

```python
good = Patch(color="forestgreen", label="Good")
ok = Patch(color="orange", label="Ok")
bad = Patch(color="crimson", label="Bad")
```

```python
turkey.plot(y="AQI Value", x="City", kind="barh", figsize=(10,50), color=colors)
plt.legend(handles=[good, ok, bad])
plt.show()
```
